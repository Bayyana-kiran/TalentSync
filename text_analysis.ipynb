```python
import re

def preprocess_text(text):
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Remove unwanted characters (replace with whitespace)
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
openness['TEXT']= openness['TEXT'].apply(lambda x: preprocess_text(x))
conscientious['TEXT']= conscientious['TEXT'].apply(lambda x: preprocess_text(x))
Agreeable['TEXT']= Agreeable['TEXT'].apply(lambda x: preprocess_text(x))
neurotic['TEXT']= neurotic['TEXT'].apply(lambda x: preprocess_text(x))
extrovert['TEXT']= extrovert['TEXT'].apply(lambda x: preprocess_text(x))
```

```python
def remove_numbers(tokens):
    return [ token for token in tokens if not token.isdigit()]

openness['TEXT_tokens']=openness['TEXT_tokens'].apply(remove_numbers)
conscientious['TEXT_tokens']=conscientious['TEXT_tokens'].apply(remove_numbers)
Agreeable['TEXT_tokens']=Agreeable['TEXT_tokens'].apply(remove_numbers)
neurotic['TEXT_tokens']=neurotic['TEXT_tokens'].apply(remove_numbers)
extrovert['TEXT_tokens']=extrovert['TEXT_tokens'].apply(remove_numbers)
```

```python
import string

def remove_punctuations(tokens):
    return [ token for token in tokens if not token in string.punctuation ]

openness['TEXT_tokens']=openness['TEXT_tokens'].apply(remove_punctuations)
conscientious['TEXT_tokens']=conscientious['TEXT_tokens'].apply(remove_punctuations)
Agreeable['TEXT_tokens']=Agreeable['TEXT_tokens'].apply(remove_punctuations)
neurotic['TEXT_tokens']=neurotic['TEXT_tokens'].apply(remove_punctuations)
extrovert['TEXT_tokens']=extrovert['TEXT_tokens'].apply(remove_punctuations)
```

```python
def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    return filtered_tokens

openness['TEXT_tokens']=openness['TEXT_tokens'].apply(remove_stopwords)
conscientious['TEXT_tokens']=conscientious['TEXT_tokens'].apply(remove_stopwords)
Agreeable['TEXT_tokens']=Agreeable['TEXT_tokens'].apply(remove_stopwords)
neurotic['TEXT_tokens']=neurotic['TEXT_tokens'].apply(remove_stopwords)
extrovert['TEXT_tokens']=extrovert['TEXT_tokens'].apply(remove_stopwords)
```

```python
def document_vector(tokens, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    num_words = 0

    for word in tokens:
        if word in model.wv:
            num_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])

    if num_words > 0:
        feature_vector = np.divide(feature_vector, num_words)
    return feature_vector
num_features = 100
document_vectors = []
```

```python
for tokens in preprocessed_text:
    vector = document_vector(tokens, model, num_features)
    document_vectors.append(vector)
```

```python
def preprocess_and_vectorize(text):
    lower = preprocess_text(text)
    tokens = lower.apply(lambda text: word_tokenize(text))
    removed_num = tokens.apply(remove_numbers)
    remove_punc = removed_num.apply(remove_punctuations)
    no_stop = remove_punc.apply(remove_stopwords)
    lemmatized = no_stop.apply(lemmatize_tokens)
    
    # Vectorization
    preprocessed_text = lemmatized
    model = Word2Vec(sentences=preprocessed_text, vector_size=100, window=5, min_count=1, sg=0)
    num_features = 100
    document_vectors = []
    
    for tokens in preprocessed_text:
        vector = document_vector(tokens, model, num_features)
        document_vectors.append(vector)
    
    return document_vectors
```

```python
def document_vector(tokens, model, num_features):
    feature_vector = np.zeros((num_features,), dtype="float32")
    num_words = 0

    for word in tokens:
        if word in model.wv:
            num_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])

    if num_words > 0:
        feature_vector = np.divide(feature_vector, num_words)
    return feature_vector
```
