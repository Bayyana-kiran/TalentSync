```python
class EntityGenerator(object):
    
    __slots__ = ['text']
    
    def __init__(self, text=None):
        self.text = text
        
    def get(self):
        nlp = spacy.load("en_core_web_sm")
        doc = nlp(self.text)
        
        skills = []
        for ent in doc.ents:
            if ent.label_ in ["GPE", "NORP", "ORG", "PERSON", "PRODUCT", "LOC"]:
                skills.append(ent.text)

        return skills
    
    def get_common_skills(self, all_skills):
        resume_skills = set(self.get())
        common_skills = list(resume_skills.intersection(all_skills))
        return common_skills

```

```python
class Resume(object):
    def __init__(self, filename=None):
        self.filename = filename
        
    def get(self):
        fFileObj = open(self.filename, 'rb')
        pdfReader = PyPDF2.PdfReader(fFileObj)
        pageObj = pdfReader.pages[0]
        print("Total Pages: {}".format(len(pdfReader.pages)))

        resume = pageObj.extract_text()
        return resume
```

```python

def analyze_technologies(common_skills, technology_skills):
    matched_technologies = {}

    for technology, skills in technology_skills.items():
        matched_skills = set(skills).intersection(common_skills)
        if matched_skills:
            matched_technologies[technology] = list(matched_skills)

    return matched_technologies
```

```python
matched_technologies = analyze_technologies(common_skills, technology_skills)

print("Common Skills:", common_skills)
print("Matched Technologies:")
for technology, skills in matched_technologies.items():
    print(f"{technology}: {', '.join(skills)}")
```
